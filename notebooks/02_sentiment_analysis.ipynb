# Etapa 2 — Pipeline de Sentimento: PETR4

Validação visual da coleta de notícias (`feedparser`) e da classificação de sentimento (`FinBERT-PT-BR`).

```python
from quantumfinance.data.news_data import fetch_news
from quantumfinance.features.sentiment import aggregate_sentiment, classify_sentiment
```

```python
news = fetch_news("PETR4")
print(f"{len(news)} notícia(s) encontrada(s)")
news
```

```python
for item in news:
    text = f"{item['title']} {item.get('summary', '')}"
    result = classify_sentiment(text)
    print(f"[{result['label']}] (score={result['scores'][result['label']]}) {item['title']}")
```
