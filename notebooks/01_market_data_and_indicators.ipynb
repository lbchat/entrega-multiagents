# Etapa 1 — Pipeline de Mercado: PETR4

Validação visual da coleta de dados (`yfinance`) e dos indicadores técnicos (`pandas-ta`).

```python
import matplotlib.pyplot as plt
import pandas_ta as ta

from quantumfinance.data.market_data import fetch_ohlcv
from quantumfinance.features.technical import calculate_indicators
```

```python
df = fetch_ohlcv("PETR4")
df["sma_20"] = ta.sma(df["Close"], length=20)
df["sma_50"] = ta.sma(df["Close"], length=50)
df["rsi"] = ta.rsi(df["Close"], length=14)
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(df.index, df["Close"], label="Close")
ax.plot(df.index, df["sma_20"], label="SMA 20")
ax.plot(df.index, df["sma_50"], label="SMA 50")
ax.set_title("PETR4 — Preço de fechamento e médias móveis")
ax.set_xlabel("Data")
ax.set_ylabel("Preço (R$)")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(df.index, df["rsi"], label="RSI 14", color="purple")
ax.axhline(70, color="red", linestyle="--", linewidth=1)
ax.axhline(30, color="green", linestyle="--", linewidth=1)
ax.set_title("PETR4 — RSI (14)")
ax.set_xlabel("Data")
ax.set_ylabel("RSI")
ax.legend()
plt.show()
```
